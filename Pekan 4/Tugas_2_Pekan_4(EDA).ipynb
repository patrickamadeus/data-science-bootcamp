Jawaban:
- Univariate Analysis merupakan sebuah teknik dalam memahami dan melakukan eksplorasi data. Imbuhan awal ‘Uni’ artinya ‘satu’, maka analisa univariate merupakan analisa data feature tunggal.
-   Bivariate analysis digunakan untuk untuk menganalisa 2 variables dan menemukan sebuah relasi. Analisis bivariate juga merupakan salah satu cara untuk menggunakan koefisien korelasi dalam rangka menemukan apakah dua variabel memiliki relasi atau tidak. 
-  Multivariate analysis digunakan untuk menganalisa lebih dari 2 variabel di waktu yang sama, trends yang dihasilkan bisa menjadi multidimensi secara alami, dengan analisa ini akan membantu kita memahami manakah data yang memiliki tren yang kompleks pada kombinasi atribut.

# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()



```python
#code here
df2 = df.dropna(axis = 0)

df2.groupby('species').mean()
```

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
*   Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g
*   Jelaskan insight yang diperoleh dari ke dua grafik tersebut.





```python
#code here 
df2.groupby('species').mean().plot(kind = "pie", y ="flipper_length_mm" , autopct = "%.1f%%" ,radius = 2)
plt.ylabel("")
plt.legend().remove()
plt.show()
```

```python
f,ax = plt.subplots(figsize=(10,8))
df['body_mass_g'].hist()
plt.xlabel("Body_mass_g")
plt.ylabel("Count")
plt.grid(False)
plt.show()
```

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
#code here
sns.catplot(data = df, x = 'sex' , y = 'bill_length_mm' , kind = "bar")
plt.show()
```

```python
f,ax = plt.subplots(figsize = (12,8))
sns.scatterplot(data = df , x = 'body_mass_g' , y = 'flipper_length_mm' ,hue = 'bill_length_mm')
plt.show()
```

- Dari scatterplot ,  dapat dilihat hubungan sumbu x dan sumbu y cenderung berbanding lurus. Peningkatan body mass cenderung menyebabkan peningkatan flipper length begitu juga sebaliknya. # Soal 5. Multivariate Analisis

*   Dengan menggunakan seaborn, buatlah sebuah pairplot keseluruhan data feature pada dataset (df) dengan argumen kind='bar', seperti expected di bawah:
*   Jelaskan hasil analisa dari multivariate yang dihasilkan



```python
#code here
sns.pairplot( data  = df)
plt.show()
```
