```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig , ax = plt.subplots(nrows=1 , ncols=3 ,squeeze=False,sharex = True,sharey = True , figsize = (18,8))
ax[0][0].plot(x,tan,label = "tanX" , color = "blue")
ax[0][1].plot(x,sin , label = "sinX" , color = "red")
ax[0][2].plot(x,cos , label = "cosX" , color = "green")
ax[0][0].set_title("Grafik tanx")
ax[0][1].set_title("Grafik sinx")
ax[0][2].set_title("Grafik cosx")
fig.legend(loc = "upper center")
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code 
fig,ax = plt.subplots(figsize= (12,7))
ax.plot(x,y)

ax1 = ax.inset_axes([-0.1,0.9,0.3,0.3])
ax1.plot(x,y,color = "blue")
ax1.plot(x,y5 , color = "y")
ax1.set_title("Grafik CosX^5")
ax1.set_xticks([-6,-4,-2,0,2,4,6])

ax2 = ax.inset_axes([0.9,0.9,0.3,0.3])
ax2.plot(x,y , color = "blue")
ax2.plot(x,y4 , color = "black")
ax2.set_title("Grafik CosX^4")
ax2.set_xticks([-6,-4,-2,0,2,4,6])

ax3 = ax.inset_axes([-0.1,-0.1,0.3,0.3])
ax3.plot(x,y , color = "blue")
ax3.plot(x,y2 , color = "red")
ax3.set_title("Grafik CosX^2")
ax3.set_xticks([-6,-4,-2,0,2,4,6])

ax4 = ax.inset_axes([0.9,-0.1,0.3,0.3])
ax4.plot(x,y , color = "blue")
ax4.plot(x,y3 , color = "green")
ax4.set_title("Grafik CosX^3")
ax4.set_xticks([-6,-4,-2,0,2,4,6])

plt.title("Grafik CosX")
plt.show()

```
