Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
plt.figure(figsize = (10,7))
plt.plot(x,cos,color = "r", label = "Cosx")
plt.plot(x,cos3,label = "3 Cosx", linestyle = "--" , color = "green")
plt.plot(x,sin,label = "Sin(X-90)" , linestyle="-.",color = "blue")
plt.plot(x,sin2,label="2 Sinx" , marker = "x",linestyle = " ",color = "yellow")
plt.legend(loc= "center" , framealpha = 0.5)
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Grafik Cos Sin")
plt.show()
```
