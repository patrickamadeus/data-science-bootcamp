```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
df_group = df.groupby('Genre').mean()[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df_group
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
import seaborn as sns
sns.set_theme()
df_group.transpose().plot(kind="bar" , figsize = (20,10))
plt.title("Mean Sales Video Games by Genre" , fontsize = 15)
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.xticks(rotation =  360)
plt.legend(loc = (0.95,0.78))
plt.show()
```

Jawab: 
Insight yang didapat adalah sbb. :
- Untuk Region North America penjualan tertinggi ada pada genre game Shooter dan Platform dan cenderung rendah untuk Strategy dan Adventure
- Untuk region Europe penjualan serupa yaitu Shooter dan Platform meski lebih tinggi untuk Shooter game. Cenderung rendah untuk Strategy dan Adventure
- Untuk region Japan, genre Role-Playing kokoh menjadi genre favorit berkebalikan dengan dua region sebelumnya, genre Shooter kurang diminati di region ini
- Untuk Region lainnya, penjualan cukup seimbang kecuali untuk genre Shooter dan Adventure yang juga cenderung rendah

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
df_group.transpose().plot(kind="bar" , figsize = (20,10) , stacked = True)
plt.title("Mean Sales Video Games by Genre" , fontsize = 15)
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.xticks(rotation =  360)
plt.legend(loc = (0.95,0.78))
plt.show()
```

Jawab:
Insight yang didapat sebenarnya serupa dengan soal sebelumnya, tetapi dengan visualisasi area dapat mempermudah penjelasannya :
- Dari kecenderungan area sub-bar di stacked bar chart NA dan EU, dapat dilihat warna ungu mendominasi dan juga warna kuning tua sehingga dapat disimpulkan penjualan terbesar ada di genre platform dan shooter. Begitu pula dengan porsi kecil di kedua warna orange (strategy adventure) 
- Untuk daerah JP , kita juga melihat persebaran area nya , kali ini didominasi oleh warna abu-abu yakni Role-Playing genre.
- Untuk other_Sales , konsisten dengan pengamatan di nomor sebelumnya diamna penjualan cenderung seimbang kecuali untuk beberapa genre.
